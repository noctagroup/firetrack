```python
# Dropdown com órbitas/pontos
dropdown_orbitaponto = widgets.Dropdown(
    options=sorted(data["candidates"].keys()),
    description="Órbita/Ponto:"
)

def render_images(orbita_ponto):
    clear_output(wait=True)
    display(dropdown_orbitaponto)
    
    imagens = data["candidates"][orbita_ponto]
    widgets_list = []

    for idx, img in enumerate(imagens):
        img_box = widgets.VBox()
        
        img_html = f"""
        <b>ID:</b> {img['id']}<br>
        <b>Data:</b> {img['datetime']}<br>
        <img src="{img['thumb']}" width="400"><br>
        <b>Status atual:</b> {"✅ Válido" if img['valid'] is True else "❌ Inválido" if img['valid'] is False else "❔ Não avaliado"}
        """
        html_widget = widgets.HTML(value=img_html)

        btn_valido = widgets.Button(description="✅ Válido", button_style="success")
        btn_invalido = widgets.Button(description="❌ Inválido", button_style="danger")
        
        def marcar_valido(b, i=idx):
            data["candidates"][orbita_ponto][i]["valid"] = True
            render_images(orbita_ponto)

        def marcar_invalido(b, i=idx):
            data["candidates"][orbita_ponto][i]["valid"] = False
            render_images(orbita_ponto)

        btn_valido.on_click(marcar_valido)
        btn_invalido.on_click(marcar_invalido)
        
        img_box.children = [html_widget, widgets.HBox([btn_valido, btn_invalido])]
        widgets_list.append(img_box)

    display(widgets.VBox(widgets_list))

# Evento ao trocar órbita/ponto
dropdown_orbitaponto.observe(lambda change: render_images(change["new"]), names="value")

# Exibir interface inicial
display(dropdown_orbitaponto)
render_images(dropdown_orbitaponto.value)
```
